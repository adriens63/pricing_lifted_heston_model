# Implied volatility in the Lifted Heston model.

Consider a Lifted Heston model as defined in slide 12 of Part III of the lectures on rough volatility. Consider an initial curve $g_0(t)$ given by
$$
g_0(t)=V_0+\lambda \theta \sum_{i=1}^n c_i \int_0^t \mathrm{e}^{-x_i(t-s)} \mathrm{d} s
$$
The weights $\left(c_i\right)_{i=1}^n$ and the means reversions $\left(x_i\right)_{i=1}^n$ are parametrized as in slide 10 of Part III of the lectures on rough volatility with a constant $r$ which we will denote by $r_n$. Define the process
$$
M_t=\exp \left(u \log \left(S_t\right)+\phi(t, T)+\sum_{i=1}^n c_i \psi^i(T-t) U_t^i\right)
$$
where $u \in \mathbb{C}$, and $\phi, \psi^i$ are functions with $\psi^i(0)=0, i=1, \ldots, n$, that satisfy the Ordinary Differential Equations (ODEs) described in slide 14 of Part III of the lectures on rough volatility.
(a) Deduce by Itô's formula that $M$ is a local martingale.
(b) Show that if $M$ is a true martingale then
$$
M_t=\mathbb{E}\left[\exp \left(u \log S_T\right) \mid \mathcal{F}_t\right] .
$$


### Let us deduce by Itô’s formula that M is a local martingale.

Since the exponential function  is regular enough we can apply Ito's formula on $M_t$ with respect to $t$, $S_t$ and $U^i_t$:

$$
df(t,S_t,U_t)= \frac{\partial f}{\partial t}(t, S_t, U_t)dt + \frac{\partial f}{\partial x}(t, S_t, U_t)dS_t + \sum_{i=1}^n \frac{\partial f}{\partial y}(t, S_t, U_t)dU^i_t +  \frac{1}{2}\frac{\partial f^2}{\partial^2 x}(t, S_t, U_t)d\left\langle S_{\cdot}\right\rangle_t +   \sum_{i=1}^n \frac{1}{2}\frac{\partial f^2}{\partial^2 y}(t, S_t, U_t)d\left\langle U^i_{\cdot}\right\rangle_t
+  \sum_{i=1}^n  \frac{\partial f^2}{\partial x \partial y}(t, S_t, U_t)d\left\langle S_{\cdot}; U^i_{\cdot}\right\rangle_t
$$

We know that : 

\begin{cases}
dS_t =  S_t\sqrt{V_t}dB_t\\
dU^i_t = (-x_iU^i_t - \lambda V_t) dt + \nu \sqrt{V_t}dW_t\\ 
\end{cases}

and :

\begin{cases}
d\left\langle S_{\cdot}\right\rangle_t =  S_t^2V_tdt\\
d\left\langle U^i_{\cdot}\right\rangle_t = \nu^2V_tdt\\ 
\end{cases}

Since the 2 brownian motions are correlated by $\rho$ we can deduce that : 
$$
d\left\langle S_{\cdot};U^i_{\cdot}\right\rangle_t = S_tV_t\nu\rho dt \\
$$

Pugging in all these terms we find : 


$$
\frac{dM_t }{M_t} =  \underbrace{\phi'(t,T)dt - \sum_{i=1}^n (\psi^i(T-t))'U^i_tc_i}_{\partial_t}dt + \underbrace{\sum_{i=1}^n \psi^i(T-t)c_i(-x_iU^i_t - \lambda V_t)}_{\partial_y}dt + \underbrace{\frac{1}{2}\frac{(u^2 - u)\cancel{S_t^2}V_t}{\cancel{S_t^2}}}_{\partial_{xx}}dt \\ + \underbrace{\frac{1}{2}\left(\sum_{i=1}^n c_i\psi^i(T-t)\right)^2 \nu^2 V_t dt}_{\partial_{yy}} + \underbrace{u\rho \frac{\cancel{S_t} V_t}{\cancel{S_t}} \nu \sum_{i=1}^n \psi^i(T-t)c_i dt}_{\partial_{xy}} + \underbrace{u\frac{1}{\cancel{S_t}}\cancel{S_t}\sqrt{V_t}dB_t}_{\partial_x} + \underbrace{\sum_{i=1}^n \psi^i(T-t)c_i \nu \sqrt{V_t}dW_t}_{\partial_y} \\
$$

We also recall that : $V_t = g_0(t) + \sum_{i=1}^n U^i_tc_i$.

To prove that M_t is a local martingale we must show that the term in $dt$ is null. Let us focus on this term we are calling "$a$" for simplicity : 

We factorise the expression by $V_t$ for some terms : 

$$
a = \phi'(t,T) - \sum_{i=1}^n (\psi^i(T-t))'U^i_tc_i - \sum_{i=1}^n \psi^i(T-t)c_ix_iU^i_t \\ + \underbrace{\left(g_0(t) + \sum_{i=1}^n U^i_tc_i\right)}_{=V_t} \left[ -\lambda \xi + \frac{u^2 - u}{2} + \frac{1}{2}\xi \nu^2 + u\rho  \nu \xi   \right]
$$

with $\xi = \sum_{i=1}^{n} c_i\psi^i(T-t) $ 

$$
\Leftrightarrow a = \phi'(t,T) - \sum_{i=1}^n (\psi^i(T-t))'U^i_tc_i - \sum_{i=1}^n \psi^i(T-t)c_ix_iU^i_t \\ + \underbrace{\left(g_0(t) + \sum_{i=1}^n U^i_tc_i\right)}_{=V_t} \left[ -\lambda \xi + \frac{u^2 - u}{2} + \frac{1}{2}\xi \nu^2 V_t dt + u\rho  \nu \xi   \right]
$$

Moreover, we know that $(\psi^i)_{i=1}^n$ solves the system of Ricatti ODEs so we consider as admitted the following equations : 

\begin{cases}
\left(\psi^i\right)^{\prime}=-x_i \psi^i+F\left(u, \sum_{j=1}^n c_j \psi^j\right) ; \quad \psi^i(0)=0 \\
F(u, \xi)=\frac{1}{2}\left(u^2-u\right)+(\rho \nu u-\lambda) \xi+\frac{\nu^2}{2} \xi^2 \\
\phi(t, T)=\int_0^{T-t} F\left(u, \sum_{j=1}^n c_j \psi^j(s)\right) g_0(T-s) \mathrm{d} s
\end{cases}

Hence we can write : $\phi'(t,T) = - F(u,v)g_0(t)$


thus :
$$
a = \underbrace{- F(u,\xi)g_0(t)}_{=\phi'(t,T)} - \sum_{i=1}^n \underbrace{\left(-x_i \psi^i(T-t)+F\left(u, \xi \right) \right)}_{=\left(\psi^i\right)^{\prime}}U^i_tc_i - \sum_{i=1}^n \psi^i(T-t)c_ix_iU^i_t \\ + \underbrace{\left(g_0(t) + \sum_{i=1}^n U^i_tc_i\right)}_{=V_t} \left[F(u, \xi) \right]
$$

We now gather th terms together and so solution appears : 


$$
a = \color{green}{\cancel{- F(u,\xi)g_0(t)}} -  \color{red}{\cancel{F\left(u, \xi \right)\sum_{i=1}^n  U^i_tc_i}} +  \color{blue}{\cancel{\sum_{i=1}^n \left(-x_i \psi^i(T-t)U^i_tc_i \right)}} - \color{blue}{\cancel{\sum_{i=1}^n \left(-x_i \psi^i(T-t)U^i_tc_i \right)}}  + \color{green}{\cancel{F(u,\xi)g_0(t)}} + \color{red}{\cancel{F\left(u, \xi \right)\sum_{i=1}^n  U^i_tc_i}} = 0 
$$

So we can conclude that $M_t$ is a **Local Martingale**.

### Since M is a true martingale then let us show : 
$$
M_t=\mathbb{E}\left[\exp \left(u \log S_T\right) \mid \mathcal{F}_t\right].
$$



Knowing that in $t = T$ : $ \psi^i(T-t)= \psi^i(0) = 0 $ and $ \phi(t, T) = \phi(T, T) = \int_0^{0} F\left(u, \sum_{j=1}^n c_j \psi^j(s)\right) g_0(T-s) \mathrm{d} s = 0 $ so we have only one term left in the exponential : 

$$
M_t=e^{u log(S_t)}
$$

Moreover if $M_t$ is a **True martingale** then by definition of a true martingale we have : 

$$
M_t=\mathbb{E}\left[\exp \left(u \log S_T\right) \mid \mathcal{F}_t\right]
$$

## Tools.

```python
eps = 1e-4
```

```python
def complex_quad(func, a, b, **kwargs):
    def real_func(x):
        return np.real(func(x))

    def imag_func(x):
        return np.imag(func(x))

    real_integral = quad(real_func, a, b, **kwargs, epsabs=eps, epsrel=eps)
    imag_integral = quad(imag_func, a, b, **kwargs)
    return (
        real_integral[0] + 1j * imag_integral[0],
        real_integral[1:],
        imag_integral[1:],
    )
```

## Parameters.

```python
T = 1.0
L = 100

S0 = 1
rho = -0.7
H = 0.1
alpha = 0.6
lamb = 0.3
alpha2 = 1
theta = 0.02
nu = 0.3
V0 = 0.02
M = 100
```

```python
log_strikes = np.linspace(-1.2, 0.2, num=20)
```

```python
num_factors = [5, 10, 20, 50]
```

```python
def ch_lifted_heston(u, S0, T, rho, lamb, theta, nu, V0, n, rn, alpha, M):
    # def F.
    F = (
        lambda u_, v: 0.5 * (u_**2 - u_)
        + (rho * nu * u_ - lamb) * v
        + 0.5 * (nu * v) ** 2
    )

    u *= 1j

    # def c_i and x_i.
    c = np.array(
        [rn ** (i - n / 2) for i in range(n)]
    )  # not 'i-1' in this formula because we use python convention that starts at index 0.
    x = copy.deepcopy(c)
    c **= 1 - alpha
    c *= (rn ** (1 - alpha) - 1) / (gamma(alpha) * gamma(2 - alpha))
    x *= ((1 - alpha) * (rn ** (2 - alpha) - 1)) / (
        (2 - alpha) * (rn ** (1 - alpha) - 1)
    )

    # solve the Riccati ODEs.
    t_abs = np.linspace(
        0, T, num=M
    )  # array of abscice [0., delta_t, 2 * delta_t, ... , (M-1) * delta_t] ~ [0, T].
    delta_t = t_abs[1]  # since t[0] is 0 by def.
    psis = np.zeros(
        [n, t_abs.shape[0]], dtype="complex_"
    )  # contains all of the psis functions. A row is a function.
    factor = 1 / (1 + x * delta_t)
    for idx in range(
        1, len(t_abs)
    ):  # the inititalization is already done by construction, so one starts at index 1.
        psis[:, idx] = factor * (
            psis[:, idx - 1] + F(u, np.dot(c, psis[:, idx - 1])) * delta_t
        )

    # def g_0.
    g_0 = V0 + lamb * theta * (
        np.expand_dims(c, axis=0)
        * (
            (1 - np.exp(-np.expand_dims(x, axis=0) * t_abs.reshape(-1, 1)))
            / np.expand_dims(x, axis=0)
        )
    ).sum(axis=-1)

    integrand_fn = (
        lambda s: F(u, np.dot(c, psis[:, floor(s / delta_t)]))
        * g_0[floor(T - s / delta_t)]
    )
    integrand = np.vectorize(integrand_fn)(t_abs)
    return np.exp(u * np.log(S0) + trapezoid(integrand, t_abs))
```

```python
def ch_lifted_heston_quad(u, S0, T, rho, lamb, theta, nu, V0, n, rn, alpha, M):
    # def F.
    F = (
        lambda u_, v: 0.5 * (u_**2 - u_)
        + (rho * nu * u_ - lamb) * v
        + 0.5 * (nu * v) ** 2
    )

    u *= 1j

    # def c_i and x_i.
    c = np.array(
        [rn ** (i - n / 2) for i in range(n)]
    )  # not 'i-1' in this formula because we use python convention that starts at index 0.
    x = copy.deepcopy(c)
    c **= 1 - alpha
    c *= (rn ** (1 - alpha) - 1) / (gamma(alpha) * gamma(2 - alpha))
    x *= ((1 - alpha) * (rn ** (2 - alpha) - 1)) / (
        (2 - alpha) * (rn ** (1 - alpha) - 1)
    )

    # solve the Riccati ODEs.
    t_abs = np.linspace(
        0, T, num=M + 1
    )  # array of abscice [0., delta_t, 2 * delta_t, ... , (M-1) * delta_t] ~ [0, T].
    delta_t = t_abs[1]  # since t[0] is 0 by def.
    psis = np.zeros(
        [n, t_abs.shape[0]], dtype="complex_"
    )  # contains all of the psis functions. A row is a function.
    factor = 1 / (1 + x * delta_t)
    for idx in range(
        1, len(t_abs)
    ):  # the inititalization is already done by construction, so one starts at index 1.
        psis[:, idx] = factor * (
            psis[:, idx - 1] + F(u, np.dot(c, psis[:, idx - 1])) * delta_t
        )

    # def g_0.
    g_0 = V0 + lamb * theta * (
        np.expand_dims(c, axis=0)
        * (
            (1 - np.exp(-np.expand_dims(x, axis=0) * t_abs.reshape(-1, 1)))
            / np.expand_dims(x, axis=0)
        )
    ).sum(axis=-1)

    return np.exp(
        u * np.log(S0)
        + complex_quad(
            lambda s: F(u, np.dot(c, psis[:, floor(s / delta_t)]))
            * g_0[floor((T - s) / delta_t)],
            0,
            T,
        )[0]
    )
```

## Call price under this model.

```python
def call_price_lifted_heston(
    S0,
    K,
    T,
    rho,
    lamb,
    theta,
    nu,
    V0,
    n,
    rn,
    alpha,
    M,
    alpha2,
    L,
    riemann_steps,
    r_int=0,
):
    k = np.log(K)

    N = riemann_steps

    factor = np.exp(-r_int * T - alpha2 * k) / np.pi

    def phi_T(u):
        return ch_lifted_heston(u, S0, T, rho, lamb, theta, nu, V0, n, rn, alpha, M)

    fun = (
        lambda u: phi_T(u - (alpha2 + 1) * 1j)
        * np.exp(-1j * np.log(K) * u)
        / ((alpha2 + 1j * u) * (alpha2 + 1 + 1j * u))
    )

    t_abs = np.linspace(0, L, N)
    integrand = np.vectorize(fun)(t_abs)

    return factor * np.real(trapezoid(integrand, t_abs))
```

```python
def call_price_lifted_heston_quad(
    S0, K, T, rho, lamb, theta, nu, V0, n, rn, alpha, M, alpha2, L, r_int=0
):
    factor = np.exp(-r_int * T - alpha2 * np.log(K)) / np.pi

    def phi_T(u):
        return ch_lifted_heston(u, S0, T, rho, lamb, theta, nu, V0, n, rn, alpha, M)

    def fun(u):
        return np.real(
            phi_T(u - (alpha2 + 1) * 1j)
            * np.exp(-1j * np.log(K) * u)
            / ((alpha2 + 1j * u) * (alpha2 + 1 + 1j * u))
        )

    return (
        factor
        * quad(
            fun,
            0,
            L,
        )[0]
    )
```

## Timer to find the best method.

```python
import timeit


class CodeTimer:
    def __init__(self, name=None, num_iter=None):
        self.name = " '" + name + "'" if name else ""
        self.num_iter = num_iter

    def __enter__(self):
        self.start = timeit.default_timer()

    def __exit__(self, exc_type, exc_value, traceback):
        num_iter = float(self.num_iter) if self.num_iter is not None else 1.0
        self.took = (timeit.default_timer() - self.start) * 1000.0 / num_iter
        print("Code block" + self.name + " took: " + str(self.took) + " ms")
```

```python
def bs_call(S0, K, T, r, vol):
    d1 = (np.log(S0 / K) + (r + 0.5 * vol**2) * T) / (vol * np.sqrt(T))
    d2 = d1 - vol * np.sqrt(T)
    Phid1 = norm.cdf(d1)
    Phid2 = norm.cdf(d2)
    call = S0 * Phid1 - K * np.exp(-r * T) * Phid2
    return call


def bs_vega(S0, K, T, r, vol):
    d1 = (np.log(S0 / K) + (r + 0.5 * vol**2) * T) / (vol * np.sqrt(T))
    return S0 * np.sqrt(T) * np.exp(-0.5 * d1**2) / np.sqrt(2 * np.pi)


def find_vol(target_value, S, T, K, r, precision, *args):
    MAX_ITERATIONS = 200
    sigma = 0.5
    for i in range(0, MAX_ITERATIONS):
        price = bs_call(S, K, T, r, sigma)
        vega = bs_vega(S, K, T, r, sigma)
        diff = target_value - price  # our root
        if abs(diff) < precision:
            return sigma
        sigma = sigma + diff / vega  # f(x) / f'(x)
    return sigma  # value wasn't found, return best guess so far
```

**NB :** *implicit volatility calculated numerically to avoid instability issues of existing closed formulas for deep ITM strikes*

```python
def approximated_implied_vol_bs_Hallerbach(
    C: float, S: float, T: float, K: float, r: float = 0
) -> float:
    X = K * np.exp(-r * T)
    f1 = np.sqrt(2 * np.pi) / (2 * (S + X))
    f2 = 2 * C + X - S
    f3 = f2 * f2 - 1.85 * ((S + X) * (X - S) * (X - S)) / (np.pi * np.sqrt(X * S))

    return f1 * (f2 + np.sqrt(f3)) / np.sqrt(T)
```

```python
def approximated_implied_vol_bs(
    C: float, S: float, T: float, K: float, r: float = 0
) -> float:
    alpha = 4
    X = K * np.exp(-r * T)
    temp = np.sqrt(2 * np.pi) * (C - 0.5 * (S - X)) / (S + X)

    return (temp + np.sqrt(temp**2 - alpha * ((S - X) / (S + X)) ** 2)) / np.sqrt(T)
```

```python
implied_vols = np.zeros_like(call_prices)

for i, n in enumerate(tqdm(num_factors)):
    for j, k in enumerate(log_strikes):
        implied_vols[i, j] = find_vol(call_prices[i, j], S0, T, np.exp(k), 0, 1.0e-5)
```

## Other Fourrier inversion formulas to compute the call prices (Gil-Pelaez's and Lewis')

```python
def call_price_lifted_heston_gil_pelaez_inversion(
    S0,
    K,
    T,
    rho,
    lamb,
    theta,
    nu,
    V0,
    n,
    rn,
    alpha,
    M,
    alpha2,
    L,
    riemann_steps,
    r_int=0,
):
    k = np.log(K)

    N = riemann_steps

    def phi_T(u):
        return ch_lifted_heston(u, S0, T, rho, lamb, theta, nu, V0, n, rn, alpha, M)

    phi_T_minus_i = phi_T(-1j)

    def fun_1(u):
        return phi_T(u - 1j) * np.exp(-1j * u * k) / (1j * u * phi_T_minus_i)

    def fun_2(u):
        return phi_T(u) * np.exp(-1j * u * k) / (1j * u)

    t_abs = np.linspace(0, L, N)[1:]
    integrand_1 = np.vectorize(fun_1)(t_abs)
    integrand_2 = np.vectorize(fun_2)(t_abs)

    pi_1 = 0.5 + (1 / np.pi) * np.real(trapezoid(integrand_1, t_abs))
    pi_2 = 0.5 + (1 / np.pi) * np.real(trapezoid(integrand_2, t_abs))

    return S0 * pi_1 - K * np.exp(-r_int * T) * pi_2
```

```python
def call_price_lifted_heston_lewis_inversion(
    S0,
    K,
    T,
    rho,
    lamb,
    theta,
    nu,
    V0,
    n,
    rn,
    alpha,
    M,
    alpha2,
    L,
    riemann_steps,
    r_int=0,
):
    k = np.log(K)

    N = riemann_steps

    def phi_T(u):
        return ch_lifted_heston(u, S0, T, rho, lamb, theta, nu, V0, n, rn, alpha, M)

    def fun_1(u):
        return phi_T(u - 1j / 2) * np.exp(-1j * u * k) / (0.25 + u * u)

    t_abs = np.linspace(-L, L, N)
    integrand_1 = np.vectorize(fun_1)(t_abs)

    integ = np.real(trapezoid(integrand_1, t_abs))

    return S0 - np.exp(-r_int * T) * np.sqrt(K) / np.pi * integ
```

## Pricing of the calls by Monte-Carlo.

Finally let us use a Monte-Carlo approach, even though the Fourrier tranform one is much more suited using this model.

```python
def call_price_lifted_heston_mc(
    num_simulations, S0, T, rho, lamb, theta, nu, V0, n, rn, alpha, M, K, r_int=0
):
    # def F.
    F = (
        lambda u_, v: 0.5 * (u_**2 - u_)
        + (rho * nu * u_ - lamb) * v
        + 0.5 * (nu * v) ** 2
    )

    # def c_i and x_i.
    c = np.array(
        [rn ** (i - n / 2) for i in range(n)]
    )  # not 'i-1' in this formula because we use python convention that starts at index 0.
    x = copy.deepcopy(c)
    c **= 1 - alpha
    c *= (rn ** (1 - alpha) - 1) / (gamma(alpha) * gamma(2 - alpha))
    x *= ((1 - alpha) * (rn ** (2 - alpha) - 1)) / (
        (2 - alpha) * (rn ** (1 - alpha) - 1)
    )

    c = np.expand_dims(c, axis=0)
    x = np.expand_dims(x, axis=0)

    t_abs = np.linspace(
        0, T, num=M
    )  # array of abscice [0., delta_t, 2 * delta_t, ... , (M-1) * delta_t] ~ [0, T].
    delta_t = t_abs[1]  # since t_abs[0] is 0 by def.

    # def g_0.
    g_0 = V0 + lamb * theta * (c * ((1 - np.exp(-x * t_abs.reshape(-1, 1))) / x)).sum(
        axis=-1
    )

    prices = []
    for _ in range(num_simulations):
        # simulate the correlated brownian motions.
        dw = np.random.normal(scale=np.sqrt(delta_t), size=M)
        dw_aux = np.random.normal(scale=np.sqrt(delta_t), size=M)
        db = rho * dw + np.sqrt(1 - rho**2) * dw_aux

        # simulate the correlated processes.
        v = np.zeros((M + 1, 1))
        u = np.zeros((M + 1, n))
        for k in range(M):
            v[k] = g_0[k] + (c * u[k]).sum(axis=1)
            u[k + 1] = (1 / (1 + x * delta_t)) * (
                u[k] - lamb * v[k] * delta_t + nu * np.sqrt(np.maximum(0, v[k])) * dw[k]
            )

        v = v.flatten()[:-1]

        # simulate the price with interest rate r_int.
        s = np.zeros_like(t_abs)
        s[0] = S0
        for idx in range(M - 1):
            s[idx + 1] = (
                1 + r_int * delta_t + np.sqrt(np.max([0.0, v[idx]])) * db[idx]
            ) * s[idx]
        with open("output.txt", "w") as f:
            print(v, file=f)
            print(*s, sep="\n", file=f)
            print("fvsvfvv" + "\n", file=f)
            print(s[-1], file=f)
            print("\n" * 4, file=f)
        prices.append(np.max([s[-1] - K, 0.0]))

    estimate = np.exp(-r_int * T) * np.mean(prices)
    std = np.std(prices)
    print(f"estimate: {estimate} with std {std} for {num_simulations} iterations.")
    print(
        f"confidence interval at the 5 percents level: [{estimate} +/- {1.96 * std/np.sqrt(num_simulations)}]"
    )
    print(
        f"which is [{estimate - 1.96 * std/np.sqrt(num_simulations), estimate + 1.96 * std/np.sqrt(num_simulations)}]"
    )
    print()

    return estimate, std
```

## Smile of vol for the Monte Carlo approach.

```python
def save_results(array_to_save, nb_sim, M, method):
    DF = pd.DataFrame(array_to_save)
    # save the dataframe as a csv file
    DF.to_csv(method + "_" + str(nb_sim) + "_" + str(M) + ".csv")
```

```python
def compute_implied_vol_mc(marketoptionPrice, S, T, K, r, precision):
    def bs_price(sigma):
        d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
        d2 = d1 - (sigma * np.sqrt(T))
        BSprice_call = S * norm.cdf(d1, 0, 1) - K * np.exp(-r * T) * norm.cdf(d2, 0, 1)
        fx = BSprice_call - 0.01 - marketoptionPrice
        return fx

    return brentq(bs_price, 0, 10, maxiter=10000)
```
